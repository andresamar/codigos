```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.signal import convolve2d as conv2
from skimage import  color, data,io, restoration
import nibabel
import shutil
from mpl_toolkits import mplot3d
from matplotlib_scalebar.scalebar import ScaleBar
from IPython.display import display, Image, SVG, Math, YouTubeVideo
```

```python
letters_45 = ["a", "c", "d", "i"]
letters_30 = ["a_30", "c_30", "d_30", "i_30"]
titulos =["Arriba", "Centro", "Derecha", "Izquierda"]
PATH_45 = "/home/a/Tesis/tiff/651-45/"
PATH_30 = "/home/a/Tesis/tiff/651-30/"
PATH_R_45 = "/home/a/Tesis/codigos/resultados/45/"
PATH_R_30 = "/home/a/Tesis/codigos/resultados/30/"
XY, XZ, YZ =[], [], []
XY_30, XZ_30, YZ_30 =[], [], []

for j in range(4):
    XY.append(io.imread(PATH_45 +letters_45[j]+ "_XY.jpg"))
    XZ.append(io.imread(PATH_45 +letters_45[j]+ "_XZ.jpg"))
    YZ.append(io.imread(PATH_45 +letters_45[j]+ "_YZ.jpg"))    
    XY_30.append(io.imread(PATH_30 +letters_30[j]+ "_XY.jpg"))
    XZ_30.append(io.imread(PATH_30 +letters_30[j]+ "_XZ.jpg"))
    YZ_30.append(io.imread(PATH_30 +letters_30[j]+ "_YZ.jpg"))    
    
```

```python
for j in range(4):
    fig = plt.figure(figsize=(10,10))
    grid = plt.GridSpec(3, 3, hspace=0.4, wspace=0.2)
    scalebar = ScaleBar(0.16,'um') #1pixel = 0.16 um
    
   
    fig.suptitle( titulos[j])
    
    gax1 = plt.subplot(grid[0, :-1])
    gax1.set_title('Plano XY')
    gax1.set_xlabel('                                                            $X(px)$   ')
    gax1.set_ylabel('                              $Y(px)$  ')
    gax1.add_artist(scalebar )

    gax2 = plt.subplot(grid[1, :-1])
    gax2.set_title('Plano XZ')
    gax2.set_xlabel('                                                    $X(px)$  ')
    gax2.set_ylabel('                                   $Z(px)$  ')

    gax3 = plt.subplot(grid[0:1, -1])
    gax3.set_title('Plano YZ')
    gax3.set_xlabel('                                                       Y$(px)$  ')
    gax3.set_ylabel('                                    $Z(px)$ ')


    gax1.imshow(XY[j],cmap ="inferno")
    gax2.imshow(XZ[j],cmap ="inferno")
    gax3.imshow(YZ[j],cmap ="inferno")
    plt.savefig(PATH_R_45+letters_45[j]+".png")
    
    
    
    gax1.imshow(XY_30[j],cmap ="inferno")
    gax2.imshow(XZ_30[j],cmap ="inferno")
    gax3.imshow(YZ_30[j],cmap ="inferno") 
    plt.savefig(PATH_R_30+letters_30[j]+".png")
```
